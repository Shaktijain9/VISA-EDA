#### Dropping case-ID as it is not relevant anymore

```python
df.drop(columns=['case_id'], inplace=True)
```

#### Adjusting the prevailing wages to yearly amount and dropping the unit of wage column

```python
month_index = df.loc[df['unit_of_wage']=='Month','prevailing_wage'].index
week_index = df.loc[df['unit_of_wage']=='Week','prevailing_wage'].index
hour_index = df.loc[df['unit_of_wage']=='Hour','prevailing_wage'].index
```

```python
df.loc[month_index.to_list(), 'prevailing_wage'] = df.loc[month_index.to_list(), 'prevailing_wage']*12
df.loc[week_index.to_list(), 'prevailing_wage'] = df.loc[week_index.to_list(), 'prevailing_wage']*52
df.loc[hour_index.to_list(), 'prevailing_wage'] = df.loc[hour_index.to_list(), 'prevailing_wage']*8*52
```

#### Dropping unit_of_wage column as we have adjusted all prevailing_wages yearly

```python
df.drop(columns=['unit_of_wage'], inplace=True)
```
